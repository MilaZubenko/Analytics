```python
import pandas as pd
import pandasql as ps
import sqlite3
```

```python
dataframe=pd.read_excel('data/bank_account_data.xlsx')
display(dataframe)
```

```python
# Задача 1 
# Опишите имеющиеся колонки в данных - сделайте аналог describe в pandas для колонок.

table_desc=ps.sqldf('''
SELECT 
COUNT(cust_id) cust_id, COUNT(purchase) purchase, COUNT(Cheq_Bal) cheq_bal, COUNT(sav_bal) sav_bal,  COUNT(debt_bal) debt_bal,  COUNT(joint_acct) joint_acct,  COUNT(age) age FROM dataframe
UNION ALL
SELECT AVG(cust_id), AVG(purchase), AVG(Cheq_Bal), AVG(sav_bal),  AVG(debt_bal),  AVG(joint_acct),  AVG(age) FROM dataframe
UNION ALL
SELECT MAX(cust_id), MAX(purchase), MAX(Cheq_Bal), MAX(sav_bal),  MAX(debt_bal),  MAX(joint_acct),  MAX(age) FROM dataframe
UNION ALL
SELECT MIN(cust_id), MIN(purchase), MIN(Cheq_Bal), MIN(sav_bal),  MIN(debt_bal),  MIN(joint_acct),  MIN(age) FROM dataframe;
''')

print('count, mean, max, min')
display(table_desc)
```

```python
# Задача 2
# Посчитайте совокупный объем средств у клиентов в банке - сумма по всем балансам.

summa=ps.sqldf('''
SELECT 
    SUM(cheq_bal) Cheq_Bal, 
    SUM(Sav_Bal) Sav_Bal, 
    SUM(Debt_Bal) Debt_Bal
FROM dataframe;
''')
display(summa)
```

```python
# Задача 3 
# Выведите только тех клиентов, которые приобрели ипотеку (колонка mortgage).

mortgage=ps.sqldf('''
SELECT 
    * 
FROM dataframe
WHERE purchase !=0;
''')
display(mortgage)
```

```python
# Задача 4 
# Какой совокупный объем средств в среднем у клиентов, которые приобрели ипотеку и у клиентов, которые её не приобретали?

grouped_summa=ps.sqldf('''
SELECT 
    purchase, 
    SUM(Cheq_Bal) Cheq_Bal, 
    SUM(Sav_Bal) Sav_bal, 
    SUM(Debt_Bal) Debt_bal
FROM dataframe
GROUP BY purchase;
''')
display(grouped_summa)
```

```python
# Задача 5 
# Есть ли разница в среднем по каждому из типов баланса у клиентов, которые приобрели ипотеку и у клиентов, 
# которые её не приобретали?

difference=ps.sqldf('''
SELECT DISTINCT 
    (SELECT avg(cheq_bal) FROM dataframe WHERE purchase=1 GROUP BY purchase) - (SELECT avg(cheq_bal) FROM dataframe WHERE purchase=0 GROUP BY purchase) AS Cheq_dif,
    (SELECT avg(sav_bal) FROM dataframe WHERE purchase=1 GROUP BY purchase) - (SELECT avg(sav_bal) FROM dataframe WHERE purchase=0 GROUP BY purchase) AS Sav_dif,
    (SELECT avg(debt_bal) FROM dataframe WHERE purchase=1 GROUP BY purchase) - (SELECT avg(debt_bal) FROM dataframe WHERE purchase=0 GROUP BY purchase) AS Debt_dif
FROM dataframe
GROUP BY purchase;
''')
display(difference)
print('Ответ: да')
```

```python
# Задача 6 
# Выберите самого возрастного получателя ипотеки.

max_age=ps.sqldf('''
SELECT *
FROM dataframe
WHERE
age=(SELECT max(age) FROM dataframe);
''')
display(max_age)
```

```python
# Задача 7 
# Выберите клиента, который раньше всех получил ипотеку.

first_client=ps.sqldf('''
SELECT *
FROM dataframe
WHERE date=(SELECT MIN(date) FROM dataframe);
''')
display(first_client)
```

```python
# Задача 8 
#Создайте любую базу данных в sqlite3, состоящую из 5 колонок, одна из колонок обязательно должна быть типа REAL. 
# (Названия колонок любые, какие захотите)

horses = sqlite3.connect('horses.db')

horses.execute('''
CREATE TABLE horses (id INTEGER, 
name STRING, 
breed STRING,
color STRING,
speed REAL ); 
''')

horses.commit()
horses.close()

```
